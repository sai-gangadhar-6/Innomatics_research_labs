```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("/content/drive/MyDrive/Innomatics Research Labs/scores_data.csv")
data.columns = data.columns.str.strip()

#As Score column Contains the month name. We should remove the Month Name(Data Cleaning)
data["Score_number"] = data["Score"].str.split("/").str[0].astype(int)
print(data.head())

# Univariate Analysis
print("\n Batch Wise Univariate analysis :")

# Histogram of scores Column
plt.hist(data["Score_number"], bins=7)
plt.title("Score Distribution")
plt.xlabel("Score")
plt.ylabel("No of Students")
plt.show()

# Bivariate Analysis

# Boxplot of scores by batch
data.boxplot(column="Score_number", by="Batch")
plt.title("Batch Wise Scores")
plt.suptitle("")
plt.xlabel("Batch")
plt.ylabel("Score")
plt.show()

# Mean scores by batch
mean = data.groupby("Batch")["Score_number"].mean()
print("\nMean score per batch:")
print(means)

# Bar plot of means
mean.plot(kind="bar")
plt.title("Average Score per Batch")
plt.xlabel("Batch")
plt.ylabel("Average Score")
plt.show()
```

# Insights Based on Batch and Score

# Insight 1  Batch Performance Comparison

Among all the three batches, AI_ELITE_7 has the highest average score, followed by AI_ELITE_6, while AI_ELITE_4 has the lowest average score. This indicates that students in AI_ELITE_7 are performing better overall compared to the other batches.


# Insight 2  Score Distribution Patterns

The histogram and boxplot show that AI_ELITE_4 has more low-scoring students, whereas AI_ELITE_7 has more students scoring in the higher range (5 to 7). This suggests stronger academic performance and consistency in AI_ELITE_7.


# Insight 3  Variability Across Batches

The spread of scores in AI_ELITE_6 is wider, indicating mixed performance within the batch  some students perform very well, while others struggle. This shows inconsistency in learning outcomes within that batch.
